<h4>Задание №1</h4>

<i>Построить матрицу c количеством строк 10 и столбцов 2, переменные
заполняются случайными числами с равномерным распределением.
Полученные сгенерированные случайные числа представить на графике в
виде точек.</i>

<p><a href="https://numpy.org/doc/1.16/reference/generated/numpy.random.uniform.html#numpy.random.uniform">Согласно документации,</a> np.random.unoform() генерирует числа из случайного распределения с возможностью указания границ для чисел (high, low), а так же размерности (size). По умолчанию пределы генерации = [0, 1).</p>
<p>Т.к. возвращается одномерный массив, решейпим его по <a href="https://numpy.org/doc/2.0/reference/generated/numpy.reshape.html">докам</a>, передавая кортеж новой размерности.</p>

```python
import numpy as np

mtrx = np.random.uniform(size = 20).reshape((10, 2))
mtrx
```

Сгенерированные 20 точек распределяем по x = 1..20 точкам, предварительно отсортировав для наглядности

```python
import matplotlib.pyplot as plt

plt.scatter(np.linspace(0, 1, 20),
            np.sort(mtrx.reshape((1, 20))),
            color='#0fff32')

plt.plot([0, 1], [0, 1], color = '#099900', markersize = 0, linestyle = 'dotted')

plt.title('20 равномерно распределенных чисел')
plt.show()
```

<h4>Задание №2</h4>

<i>Сгенерировать 1000 случайных чисел с любым известным распределением
и построить их гистограмму, математическое ожидание, дисперсию.</i>

На этот раз возьмем нормальное распределение, с заранее известными матожиданием и дисперсией

```python
preLoc = 0
preScale = 2

mtrx2 = np.random.normal(loc = preLoc, 
                         scale = preScale, 
                         size = 1000) 

plt.hist(mtrx2, bins = 50, color = 'cyan', edgecolor = 'blue')

plt.title('1000 нормально распределенных чисел')

m = np.mean(mtrx2) # Вычисляем мат ожидание по распределению
d = np.var(mtrx2) # Вычисляем дисперсию по распределению

plt.axvline(x = m, color = 'black') # Строим экспериментальное мат ожидание
plt.axvline(x = preLoc, color = 'black', linestyle = 'dotted') # Теор мат ожидание - все теоретические значения пунктиром 

plt.axvline(x = d, color = 'red') # Дисперсии, построены на +- от нуля, т.к. для удобства мат ожидание бралось = 0
plt.axvline(x = -d, color = 'red')

plt.axvline(x = preScale ** 2, color = 'red', linestyle = 'dotted') # Теоретические значения дисперсий,
                                                                    # задаются квадратом от теор. отклонения
plt.axvline(x = - preScale ** 2, color = 'red', linestyle = 'dotted')

plt.show()
```

<h4>Задание №3</h4>

<i>Сгенерировать случайную точку, равномерно распределенную в квадрате
со стороной а</i>

```python
import matplotlib.patches as patches

a = 10

# Если хочется повводить
# a = float(input('Введите сторону квадрата'))

fig = plt.figure()
ax = fig.add_subplot()

plt.xlim(0, a + 2) # Ограничим видимость на +1 от квадрата
plt.ylim(0, a + 2)

plt.title('Равномерно распределённая точка в квадрате')

# Квадрат строим как прямоугольник со сторонами a, чуть дальше от осей
rect = patches.Rectangle((1, 1), width = a, height = a, fill = False, lw = 1, ls = 'dashed', edgecolor = 'black')
ax.add_patch(rect)

# По одной точке не особо наглядно, поэтому берём чуть побольше
# Чем мельче и насыщеннее точка - тем раньше она была сгенерирована
for i in range(1, 101):
    plt.plot(np.random.uniform(size = 1) * a + 1,
            np.random.uniform(size = 1) * a + 1,
            marker = 'o', 
            markersize = i/10,
            color = '#' + hex(i * 2)[2::] * 2 + 'ff' )

plt.show()
```
