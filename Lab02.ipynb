<h4>Задание №1</h4>
<p style='width: 60%'>Сгенерировать выборку случайных чисел размером 100 и 1000 для двух
распределений – экспоненциального и нормального. Для созданных выборок сделать
следующее:</p>

<ol style='width: 60%'>
    <li>Посчитать выборочное среднее и дисперсию, сравнить с математическим
ожиданием соответствующих распределений;</li>
    <li>Посчитать 0.5 и 0.99 квантили, сравнить с соответствующими теоретическими
значениями;</li>
    <li>Построить гистограмму распределения;</li>
    <li>Построить функцию распределения случайной величины на основе выборки (на
одном графике показать функции распределения, полученные из выборок
разного размера и теоретическую);</li>
    <li>Построить плотность распределения случайной величины на основе выборки (на
одном графике показать плотности распределения, полученные из выборок
разного размера и теоретическую).</li>
</ol>

<p style='width: 60%'>В итоге проанализировать зависимость точности аппроксимации от количества
экспериментов.</p>
<hr>

```python
import numpy as np

ALL_SCALE = 4 # Отклонение, которое будет использоваться везде

# Нормальные распределения на 100 и 1000 точек
NORM_100 = np.random.normal(scale = ALL_SCALE, size = 100)
NORM_1000 = np.random.normal(scale = ALL_SCALE, size = 1000)

# Экспоненциальные распределения на 100 и 1000 точек
EXP_100 = np.random.exponential(scale = ALL_SCALE, size = 100)
EXP_1000 = np.random.exponential(scale = ALL_SCALE, size = 1000)
```

<p style='width: 60%'>После задания все "констант" посчитаем среднее и дисперсию:</p>

```python
# Среднее для нормального распределения
tempMNorm100 = np.mean(NORM_100)
tempMNorm1000 = np.mean(NORM_1000)

# Среднее для экспоненциального распределения
tempMExp100 = np.mean(EXP_100)
tempMExp1000 = np.mean(EXP_1000)

# Дисперсия для нормального распределения
tempDNorm100 = np.var(NORM_100)
tempDNorm1000 = np.var(NORM_1000)

# Дисперсия для экспоненциального распределения
tempDExp100 = np.var(EXP_100)
tempDExp1000 = np.var(EXP_1000)
```

<p style='width: 60%'>По умолчанию, мат ожидание нормального распределения, получаемое через numpy.random.normal() равно 0.</p>
<p style='width: 60%'>Мат ожидание экспоненциального распределения равно</p>
$$M = 1/\lambda = ALL\_SCALE$$

```python
print('Мат ожидание нормального распределения = %.2f, вычисленное для 100 точек = %.2f, отклонение %.2f \n' % (0.0, tempMNorm100, abs(0 - tempMNorm100)))
print('Мат ожидание нормального распределения = %.2f, вычисленное для 1000 точек = %.2f, отклонение %.2f \n' % (0.0, tempMNorm1000, abs(0 - tempMNorm1000)))

print('Мат ожидание экспоненциального распределения = %.2f, вычисленное для 100 точек = %.2f, отклонение %.2f \n' % (ALL_SCALE, tempMExp100, abs(ALL_SCALE - tempMExp100)))
print('Мат ожидание экспоненциального распределения = %.2f, вычисленное для 1000 точек = %.2f, отклонение %.2f \n' % (ALL_SCALE, tempMExp1000, abs(ALL_SCALE - tempMExp1000)))

print('Дисперсия нормального распределения = %.2f, вычисленное для 100 точек = %.2f, отклонение %.2f \n' % (ALL_SCALE ** 2, tempDNorm100, abs(ALL_SCALE ** 2 - tempDNorm100)))
print('Дисперсия нормального распределения = %.2f, вычисленное для 1000 точек = %.2f, отклонение %.2f \n' % (ALL_SCALE ** 2, tempDNorm1000, abs(ALL_SCALE ** 2 - tempDNorm1000)))

print('Дисперсия экспоненциального распределения = %.2f, вычисленное для 100 точек = %.2f, отклонение %.2f \n' % (ALL_SCALE ** 2, tempDExp100, abs(ALL_SCALE ** 2 - tempDExp100)))
print('Дисперсия экспоненциального распределения = %.2f, вычисленное для 1000 точек = %.2f, отклонение %.2f \n' % (ALL_SCALE ** 2, tempDExp1000, abs(ALL_SCALE ** 2 - tempDExp1000)))
```

```python
# Квантили 0.5 для полученных распределений
qNorm100_5 = np.quantile(NORM_100, 0.5)
qNorm1000_5 = np.quantile(NORM_1000, 0.5)

qExp100_5 = np.quantile(EXP_100, 0.5)
qExp1000_5 = np.quantile(EXP_1000, 0.5)

# Квантили 0.99 для полученных распределений
qNorm100_9 = np.quantile(NORM_100, 0.99)
qNorm1000_9 = np.quantile(NORM_1000, 0.99)

qExp100_9 = np.quantile(EXP_100, 0.99)
qExp1000_9 = np.quantile(EXP_1000, 0.99)
```

<p style='width: 60%'>Экспериментальные квантили вычислены, теперь находим теоретические и сравниваем:</p>

```python
import math

# Квантили стандартного нормального распределения
ST_NORM_9 = 2.326
ST_NORM_5 = 0

# Теоретические квантили для нормального распределения
cNorm5 = 0.0 + ALL_SCALE * ST_NORM_5
cNorm9 = 0.0 + ALL_SCALE * ST_NORM_9

# Теоретические квантили для экспоненциального распределения
cExp5 = - ALL_SCALE * math.log(1 - 0.5)
cExp9 = - ALL_SCALE * math.log(1 - 0.99)
```

Для нормального распределения квантили считались по формуле:
$$q = M + \sigma * Z(p) $$
Для экспоненциального:
$$q = - 1 / \lambda * ln(1 - p)$$, где
$$1/\lambda = ALL\_SCALE$$

```python
print('Теоретическое значение 0.5 квантиля нормального распределения для 100 чисел = %.2f, полученное = %.2f, \n абсолютное отклонение = %.2f' % (cNorm5, qNorm100_5, abs(qNorm100_5 - cNorm5)))
print('Теоретическое значение 0.5 квантиля нормального распределения для 1000 чисел = %.2f, полученное = %.2f, \n абсолютное отклонение = %.2f' % (cNorm5, qNorm1000_5, abs(qNorm1000_5 - cNorm5)))

print('Теоретическое значение 0.99 квантиля нормального распределения для 100 чисел = %.2f, полученное = %.2f, \n абсолютное отклонение = %.2f' % (cNorm9, qNorm100_9, abs(qNorm100_9 - cNorm9)))
print('Теоретическое значение 0.99 квантиля нормального распределения для 1000 чисел = %.2f, полученное = %.2f, \n абсолютное отклонение = %.2f' % (cNorm9, qNorm1000_9, abs(qNorm1000_9 - cNorm9)))

print('Теоретическое значение 0.5 квантиля экспоненциального распределения для 100 чисел = %.2f, полученное = %.2f, \n абсолютное отклонение = %.2f' % (cExp5, qExp100_5, abs(qExp100_5 - cExp5)))
print('Теоретическое значение 0.5 квантиля экспоненциального распределения для 1000 чисел = %.2f, полученное = %.2f, \n абсолютное отклонение = %.2f' % (cExp5, qExp1000_5, abs(qExp1000_5 - cExp5)))

print('Теоретическое значение 0.99 квантиля экспоненциального распределения для 100 чисел = %.2f, полученное = %.2f, \n абсолютное отклонение = %.2f' % (cExp9, qExp100_9, abs(qExp100_9 - cExp9)))
print('Теоретическое значение 0.99 квантиля экспоненциального распределения для 1000 чисел = %.2f, полученное = %.2f, \n абсолютное отклонение = %.2f' % (cExp9, qExp1000_9, abs(qExp1000_9 - cExp9)))
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2)
fig.suptitle('Гистограммы нормальных распределений на 100 и 1000 точек')
axs[0].hist(NORM_100, bins = 11, color = 'cyan', edgecolor = 'blue')
axs[1].hist(NORM_1000, bins = 51, color = 'cyan', edgecolor = 'blue')

plt.show()
```

```python
from scipy.stats import norm
from statsmodels.distributions.empirical_distribution import ECDF

x100 = np.linspace(-15, 15, 100)
x1000 = np.linspace(-15, 15, 1000)

# Функция распеределения нормального распределения штука интересная, мы воспользуемся готовым вариантом
fTheorNorm = norm.cdf(x1000, loc = 0.0, scale = ALL_SCALE)

# Получаем распределения под наши сгенерированные точки
fExpNorm100 = ECDF(NORM_100)
fExpNorm1000 = ECDF(NORM_1000)

plt.plot(x1000, fTheorNorm, label = 'Теоретическая функция')
plt.step(x100, fExpNorm100(x100), where = 'post', label = 'Эмпирическая функция для 100т.')
plt.step(x1000, fExpNorm1000(x1000), where = 'post', label = 'Эмпирическая функция для 1000т.')

plt.title('Сравнение полученных функций распределения для нормального распределения')
plt.legend()
plt.grid()
plt.show()
```

```python
from scipy.stats import expon

x100 = np.linspace(0, 20, 100)
x1000 = np.linspace(0, 20, 1000)

# С помощью stats упростим себе жизнь в получении функции функции распределения (не опечатка)
fTheorExp = expon.cdf(x1000, loc = 0.0, scale = ALL_SCALE)

# Получаем расчитанные точки функции распределения
fExpExp100 = ECDF(EXP_100)
fExpExp1000 = ECDF(EXP_1000)

plt.plot(x1000, fTheorExp, label = 'Теоретическая функция')
plt.step(x100, fExpExp100(x100), where = 'post', label = 'Эмпирическая функция для 100т.')
plt.step(x1000, fExpExp1000(x1000), where = 'post', label = 'Эмпирическая функция для 1000т.')

plt.title('Сравнение полученных функций распределения для экспоненциального распределения')
plt.legend()
plt.grid()
plt.show()
```

```python
import seaborn as sns

# Находим плотность вероятности для нормального распределения
x = np.linspace (-15, 15, 1000)
normTheor = norm.pdf(x, loc = 0.0, scale = ALL_SCALE)

# Максимально упростим себе жизнь в расчётах
sns.kdeplot(NORM_100, label = 'Для 100 точек')
sns.kdeplot(NORM_1000, label = 'Для 1000 точек')

plt.plot(x, normTheor, label = 'Теоретические значения')

plt.grid()
plt.legend()
plt.title('Сравнение плотностей вероятности')

plt.show()
```

<h4>Задание №2</h4>
<p style='width: 60%'>Сгенерировать три выборки размера 100, 1000 и 10000 для случайных
расстояний между двумя точками, равномерно распределенные в прямоугольнике со
сторонами 10 и 30. Получить среднее значение расстояния между точками, построить
функцию распределения вероятностей и плотности вероятностей случайных
расстояний. Показать разницу между соответствующими функциями на одном
графике.</p>
<hr>

```python
a = 10
b = 30

# Объявляем функцию для нахождения расстояний между двумя равномерно распределёнными точками
# Возвращаем сразу расстояния, ибо точки сами по себе не нужны
def giveMeDistance() :
    return math.sqrt((np.random.uniform(high = a) - np.random.uniform(high = a)) ** 2 + 
            (np.random.uniform(high = b) - np.random.uniform(high = b)) ** 2)

# Объявляем массивы нулей
arr100 = np.zeros((1, 100))
arr1000 = np.zeros((1, 1000))
arr10000 = np.zeros((1, 10000))

# Заполняем массивы с помощью нашей функции
for i in range(0, 100):
    arr100[0][i] = giveMeDistance()

for i in range(0, 1000):
    arr1000[0][i] = giveMeDistance()

for i in range(0, 10000):
    arr10000[0][i] = giveMeDistance()

fig, axs = plt.subplots(3)
fig.suptitle('Гистограммы для расстояний из выборки 100, 1000, 10000 соответственно')

axs[0].hist(arr100[0], bins = 5)
axs[1].hist(arr1000[0], bins = 50)
axs[2].hist(arr10000[0], bins = 500)
plt.show()
```

```python
from scipy.stats import lognorm
from scipy.stats import gaussian_kde

# Мапим точки под графики
x100 = np.linspace(min(arr100[0]), max(arr100[0]), 100)
x1000 = np.linspace(min(arr1000[0]), max(arr1000[0]), 1000)
x10000 = np.linspace(min(arr10000[0]), max(arr10000[0]), 10000)
```

Проблема здесь заключается в следующем. Нужно как-то аккуратно аппроксимировать полученные значения, не зная что это за СВ, при этом не передавая в функции аппроксимации лишних значений, которые можно и вообще не подобрать как нужно...

```python
# Используем ядровую оценку плотности для нахождения плотности вероятностей
pdf100 = gaussian_kde(arr100[0])
pdf1000 = gaussian_kde(arr1000[0])
pdf10000 = gaussian_kde(arr10000[0])

# Полученные KDE функции мапим на наши linspace
plt.plot(x100, pdf100(x100), label = 'для 100 точек')
plt.plot(x1000, pdf1000(x1000), label = 'для 1000 точек')
plt.plot(x10000, pdf10000(x10000), label = 'для 10000 точек')

plt. title('Оценка плотностей вероятности для полученных значений с KDE')
plt.grid()
plt.legend()
plt.show()
```

Плотность вероятности "на глаз" неплохо совпала с дискретными значениями, поэтому теперь ищем функцию распределения. И так как мы до сих пор не в курсе, какому распределению и с каким параметрами это всё подчиняется, то функцию распределения ищем просто дискретно "интегрируя"

```python
# Интегрирование по местным обычаям
cdf100 = np.cumsum(pdf100(x100)) * (x100[1] - x100[0])
cdf1000 = np.cumsum(pdf1000(x1000)) * (x1000[1] - x1000[0])
cdf10000 = np.cumsum(pdf10000(x10000)) * (x10000[1] - x10000[0])

cdf100 /= cdf100[-1]
cdf1000 /= cdf1000[-1]
cdf10000 /= cdf10000[-1]

# Копипаста для построения графика с прошлого пункта
plt.plot(x100, cdf100, label = 'для 100 точек')
plt.plot(x1000, cdf1000, label = 'для 1000 точек')
plt.plot(x10000, cdf10000, label = 'для 10000 точек')

plt. title('Оценка функций распределения для полученных значений с KDE')
plt.grid()
plt.legend()
plt.show()
```

Слишком уж сильно это тянет на смещенное нормальное распределение, было бы хорошо, если бы я понял это раньше...
